```python
# Step 1: Import necessary libraries
import pandas as pd

# Step 2: Upload the file from your computer
from google.colab import files
uploaded = files.upload()

# Step 3: Load the CSV file (replace with your actual file name)
file_name = list(uploaded.keys())[0]  # automatically gets the uploaded file name
df = pd.read_csv(file_name)

# Step 4: Check first few rows
df.head()
```

```python
df.columns
```

```python
# Step 3: Separate features (X) and target (y)
X = df.drop(columns=['Credit_Limit'])
y = df['Credit_Limit']

# Step 4: Split into training and test sets (70:30 ratio)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Step 5: Train Linear Regression model with fit_intercept = False
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

model = LinearRegression(fit_intercept=False)
model.fit(X_train, y_train)

# Step 6: Predict and compute r² score on test data
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)

print("R² Score on Test Dataset:", round(r2, 3))
```

```python
import numpy as np

# Get model coefficients
coefficients = model.coef_

# Find the index of the feature with the largest absolute coefficient
max_index = np.argmax(np.abs(coefficients))

print("Index of feature with highest absolute coefficient value:", max_index)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score

# Step 1: Initialize Ridge model with given parameters
ridge_model = Ridge(solver='sag', tol=0.0005, random_state=42)

# Step 2: Train on the training data
ridge_model.fit(X_train, y_train)

# Step 3: Predict on the test data
y_pred_ridge = ridge_model.predict(X_test)

# Step 4: Compute r² score
r2_ridge = r2_score(y_test, y_pred_ridge)

print("R² Score on Test Dataset (Ridge Model):", round(r2_ridge, 3))
```

```python
# Get the intercept value from the Ridge model
intercept_value = ridge_model.intercept_

print("Intercept value:", round(intercept_value, 3))
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# Step 1: Initialize the KNN Regressor
knn_model = KNeighborsRegressor(n_neighbors=10, p=1)

# Step 2: Train on training data
knn_model.fit(X_train, y_train)

# Step 3: Predict on test data
y_pred_knn = knn_model.predict(X_test)

# Step 4: Compute RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred_knn))

print("Root Mean Squared Error (RMSE) on Test Dataset:", round(rmse, 3))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# Step 1: Initialize Decision Tree Regressor with given parameters
dt_model = DecisionTreeRegressor(
    max_depth=10,
    min_samples_split=6,
    min_samples_leaf=6,
    random_state=42
)

# Step 2: Train on training data
dt_model.fit(X_train, y_train)

# Step 3: Predict on test data
y_pred_dt = dt_model.predict(X_test)

# Step 4: Compute RMSE
rmse_dt = np.sqrt(mean_squared_error(y_test, y_pred_dt))

print("Root Mean Squared Error (RMSE) on Test Dataset:", round(rmse_dt, 3))
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score

# Step 1: Define the AdaBoost regressor
ada = AdaBoostRegressor(random_state=42)

# Step 2: Define the hyperparameter grid
param_grid = {
    'n_estimators': [10, 50, 100, 200, 500],
    'learning_rate': [0.1, 0.5, 1, 2]
}

# Step 3: Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=ada,
    param_grid=param_grid,
    cv=4,
    n_jobs=-1,
    scoring='r2'
)

# Step 4: Fit on training data
grid_search.fit(X_train, y_train)

# Step 5: Get the best estimator
best_model = grid_search.best_estimator_

print("Best Parameters:", grid_search.best_params_)

# Step 6: Predict on test data using the best model
y_pred_ada = best_model.predict(X_test)

# Step 7: Compute R² score on test data
r2_ada = r2_score(y_test, y_pred_ada)

print("R² Score on Test Dataset (Best AdaBoost Model):", round(r2_ada, 3))
```

```python
best_n_estimators = grid_search.best_params_['n_estimators']
print("Best n_estimators value:", best_n_estimators)
```
