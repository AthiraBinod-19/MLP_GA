California Housing Dataset

```python
from sklearn.datasets import fetch_california_housing
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error
```

```python
# Step 1: Load the dataset
data = fetch_california_housing()
X = data.data
y = data.target
```

```python
# Step 2: Split into train/test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
#Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
r2_lr = r2_score(y_test, y_pred_lr)
rmse_lr = np.sqrt(root_mean_squared_error(y_test, y_pred_lr))
print(rmse_lr,r2_lr)
```

```python
#AdaBoost Regressor
ada = AdaBoostRegressor(random_state=1)
ada.fit(X_train, y_train)
y_pred_ada = ada.predict(X_test)
r2_ada = r2_score(y_test, y_pred_ada)
rmse_ada = np.sqrt(root_mean_squared_error(y_test, y_pred_ada))
print(rmse_ada,r2_ada)
```

```python
#Descision Tree Regressor
dt = DecisionTreeRegressor(random_state=1)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
r2_dt = r2_score(y_test, y_pred_dt)
rmse_dt = np.sqrt(root_mean_squared_error(y_test, y_pred_dt))
print(rmse_dt,r2_dt)
```
