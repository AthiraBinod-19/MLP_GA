```python
# Step 1: Install required libraries
!pip install scikit-learn mlxtend --quiet

# Step 2: Import libraries
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.linear_model import LogisticRegression
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

# Step 3: Upload and load data
from google.colab import files
uploaded = files.upload()

df = pd.read_csv('Week3_GA_dataset.csv')

# Step 4: Clean and prepare data
df.replace('?', np.nan, inplace=True)

X = df.drop(columns=['Target'])
y = df['Target']

# Convert all feature columns to numeric
for col in X.columns:
    X[col] = pd.to_numeric(X[col], errors='coerce')

numeric_features = X.columns.tolist()  # All are numeric
categorical_features = []

# Encode target using ordinal encoding
y = OrdinalEncoder().fit_transform(y.values.reshape(-1, 1)).ravel()

# Preprocessing pipeline
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features)
])

# Preprocess data
X_processed = preprocessor.fit_transform(X)

# Step 5: Backward Sequential Feature Selection (select 2 features)
sfs_backward = SFS(LogisticRegression(),
                   k_features=2,
                   forward=False,      # <--- Backward selection here
                   floating=False,
                   scoring='accuracy',
                   cv=5,
                   n_jobs=-1)

sfs_backward = sfs_backward.fit(X_processed, y)

# Step 6: Output indices and names of selected features
selected_indices_back = list(sfs_backward.k_feature_idx_)
selected_feature_names_back = [numeric_features[i] for i in selected_indices_back]

print("Indices of two most important features selected by SFS (backward):", selected_indices_back)
print("Feature names:", selected_feature_names_back)
```
