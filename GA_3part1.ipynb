```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.feature_selection import VarianceThreshold
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.compose import ColumnTransformer
import numpy as np

# Example data (replace this with your own feature matrix)
X = pd.DataFrame({
    'Feature1': [1, 2, np.nan, 4],
    'Feature2': [5, np.nan, 7, 8],
    'Feature3': [1.2, 2.3, 3.1, 4.0],
    'Feature4': [10, 20, 30, 40],
    'Feature5': ['A', 'B', 'A', 'C']
})

# Define column groups
num_cols = ['Feature1', 'Feature2', 'Feature3', 'Feature4']
cat_cols = ['Feature5']

# Define the transformations
numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_pipeline = Pipeline([
    ('encoder', OrdinalEncoder())
])

# Combine using ColumnTransformer
preprocessor = ColumnTransformer([
    ('num', numeric_pipeline, num_cols),
    ('cat', categorical_pipeline, cat_cols)
])

# Add VarianceThreshold
pipeline = Pipeline([
    ('preprocess', preprocessor),
    ('var_thresh', VarianceThreshold(threshold=0.1))
])

# Fit and transform
X_transformed = pipeline.fit_transform(X)

# Number of remaining features
print("Number of features after pipeline:", X_transformed.shape[1])
```
